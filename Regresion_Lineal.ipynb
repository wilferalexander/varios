** 驴Qu茅 es la regresi贸n lineal y c贸mo se usa en el 谩rea de la salud?**

La regresi贸n lineal es una t茅cnica de machine learning que nos ayuda a entender y predecir la relaci贸n entre dos variables. En el 谩rea de la salud, puede ser una herramienta poderosa para predecir y analizar diversos aspectos m茅dicos.

驴C贸mo funciona?

**Datos de entrada:**
 Tienes datos hist贸ricos, como la relaci贸n entre el 铆ndice de masa corporal (IMC) y la presi贸n arterial.

**Dibuja una l铆nea:**
La regresi贸n lineal encuentra la mejor l铆nea recta que atraviesa esos puntos. Esta l铆nea se llama "l铆nea de mejor ajuste".

**Predicci贸n:**
Puedes usar esta l铆nea para predecir resultados. Por ejemplo, si sabes el IMC de una persona, puedes predecir su presi贸n arterial.
F贸rmula m谩gica: La f贸rmula de la l铆nea de regresi贸n es: Y = mX + b

Y es el valor que queremos predecir (presi贸n arterial).
X es el valor de entrada (IMC).
m es la pendiente de la l铆nea (indica cu谩nto cambia Y por cada cambio en X).
b es la intersecci贸n con el eje Y (el valor de Y cuando X es 0).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: IMC vs Presi贸n Arterial (en mm Hg)
imc = np.array([18.5, 24.0, 27.5, 30.0, 35.0]).reshape(-1, 1)
presion_arterial = np.array([120, 130, 140, 150, 160])

# Crear el modelo de regresi贸n lineal
modelo = LinearRegression()
modelo.fit(imc, presion_arterial)

# Predicci贸n
imc_nuevo = np.array([28.0]).reshape(-1, 1)
presion_predicha = modelo.predict(imc_nuevo)

print(f"Para un IMC de 28.0, la presi贸n arterial predicha es: {presion_predicha[0]:.2f} mm Hg")

# Visualizaci贸n
plt.scatter(imc, presion_arterial, color='blue')
plt.plot(imc, modelo.predict(imc), color='red')
plt.xlabel('IMC')
plt.ylabel('Presi贸n Arterial (mm Hg)')
plt.title('Regresi贸n Lineal: IMC vs Presi贸n Arterial')
plt.show()
```

Este c贸digo crea un modelo de regresi贸n lineal usando datos de ejemplo (IMC y presi贸n arterial).
 Luego, predice la presi贸n arterial para una persona con un IMC de 28.0 y visualiza la relaci贸n entre el IMC y la presi贸n arterial.
