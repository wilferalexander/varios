```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: horas estudiadas vs calificaci√≥n
horas_estudio = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
calificaciones = np.array([2, 4, 5, 4, 5])

# Crear el modelo de regresi√≥n lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predicci√≥n
horas_nuevas = np.array([6]).reshape(-1, 1)
calificacion_predicha = modelo.predict(horas_nuevas)

print(f"Si estudias 6 horas, tu calificaci√≥n predicha es: {calificacion_predicha[0]:.2f}")

# Visualizaci√≥n
plt.scatter(horas_estudio, calificaciones, color='blue')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaci√≥n')
plt.title('Regresi√≥n Lineal: Horas de Estudio vs Calificaci√≥n')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: tama√±o de la vivienda (m2) vs precio (en miles de d√≥lares)
tama√±o_vivienda = np.array([50, 60, 80, 100, 120]).reshape(-1, 1)
precio_vivienda = np.array([150, 180, 220, 260, 300])

# Crear el modelo de regresi√≥n lineal
modelo = LinearRegression()
modelo.fit(tama√±o_vivienda, precio_vivienda)

# Predicci√≥n
tama√±o_nuevo = np.array([90]).reshape(-1, 1)
precio_predicho = modelo.predict(tama√±o_nuevo)

print(f"Para una vivienda de 90 m2, el precio predicho es: ${precio_predicho[0]:.2f} mil d√≥lares")

# Visualizaci√≥n
plt.scatter(tama√±o_vivienda, precio_vivienda, color='blue')
plt.plot(tama√±o_vivienda, modelo.predict(tama√±o_vivienda), color='red')
plt.xlabel('Tama√±o de Vivienda (m2)')
plt.ylabel('Precio de Vivienda (mil d√≥lares)')
plt.title('Regresi√≥n Lineal: Tama√±o de Vivienda vs Precio')
plt.show()
```
