##  驴Qu茅 es la regresi贸n lineal y c贸mo se usa en ciencias sociales?

La regresi贸n lineal es una t茅cnica de machine learning que nos ayuda a entender y predecir la relaci贸n entre dos variables. En las ciencias sociales, puede ser una herramienta poderosa para analizar la relaci贸n entre variables como el ingreso y el nivel educativo en diferentes poblaciones.

驴C贸mo funciona?

**Datos de entrada:**
Tienes datos hist贸ricos, como el nivel educativo y el ingreso de diferentes personas.

**Dibuja una l铆nea:**
 La regresi贸n lineal encuentra la mejor l铆nea recta que atraviesa esos puntos. Esta l铆nea se llama "l铆nea de mejor ajuste".

**Predicci贸n:**
Puedes usar esta l铆nea para predecir resultados futuros. Por ejemplo, si sabes el nivel educativo de una persona, puedes predecir su ingreso.
F贸rmula m谩gica: La f贸rmula de la l铆nea de regresi贸n es: Y = mX + b

Y es el valor que queremos predecir (ingreso).
X es el valor de entrada (nivel educativo).
m es la pendiente de la l铆nea (indica cu谩nto cambia Y por cada cambio en X).
b es la intersecci贸n con el eje Y (el valor de Y cuando X es 0).


Este c贸digo crea un modelo de regresi贸n lineal usando datos de ejemplo (nivel educativo e ingreso). Luego, predice el ingreso para una persona con 22 a帽os de educaci贸n y visualiza la relaci贸n entre el nivel educativo y el ingreso.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: Nivel Educativo (a帽os de educaci贸n) vs Ingreso (en miles de d贸lares)
nivel_educativo = np.array([5, 8, 10, 12, 14, 16, 18, 20]).reshape(-1, 1)
ingreso = np.array([20, 30, 35, 40, 50, 60, 70, 80])

# Crear el modelo de regresi贸n lineal
modelo = LinearRegression()
modelo.fit(nivel_educativo, ingreso)

# Predicci贸n
nivel_educativo_nuevo = np.array([22]).reshape(-1, 1)
ingreso_predicho = modelo.predict(nivel_educativo_nuevo)

print(f"Para un nivel educativo de 22 a帽os, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil d贸lares")

# Visualizaci贸n
plt.scatter(nivel_educativo, ingreso, color='blue')
plt.plot(nivel_educativo, modelo.predict(nivel_educativo), color='red')
plt.xlabel('Nivel Educativo (a帽os)')
plt.ylabel('Ingreso (mil d贸lares)')
plt.title('Regresi贸n Lineal: Nivel Educativo vs Ingreso')
plt.show()
```

**Ejemplo considerando tambi茅n la experiencia laboral:**

Este c贸digo ampl铆a el modelo de regresi贸n lineal para incluir tanto el nivel educativo como la experiencia laboral como factores predictivos del ingreso.

```python
# Datos de ejemplo: Nivel Educativo (a帽os de educaci贸n), Experiencia Laboral (a帽os) vs Ingreso (en miles de d贸lares)
nivel_educativo = np.array([5, 8, 10, 12, 14, 16, 18, 20])
experiencia_laboral = np.array([1, 3, 5, 7, 10, 12, 15, 20])
ingreso = np.array([20, 30, 35, 40, 50, 60, 70, 80])

# Crear una matriz de caracter铆sticas
X = np.column_stack((nivel_educativo, experiencia_laboral))

# Crear el modelo de regresi贸n lineal
modelo = LinearRegression()
modelo.fit(X, ingreso)

# Predicci贸n
nuevo_educacion_experiencia = np.array([[22, 25]])
ingreso_predicho = modelo.predict(nuevo_educacion_experiencia)

print(f"Para un nivel educativo de 22 a帽os y 25 a帽os de experiencia laboral, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil d贸lares")

# Visualizaci贸n
plt.scatter(nivel_educativo, ingreso, color='blue')
plt.plot(nivel_educativo, modelo.predict(X), color='red')
plt.xlabel('Nivel Educativo (a帽os)')
plt.ylabel('Ingreso (mil d贸lares)')
plt.title('Regresi贸n Lineal: Nivel Educativo y Experiencia vs Ingreso')
plt.show()
```
