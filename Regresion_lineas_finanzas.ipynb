🔍📈 ¿Qué es la regresión lineal y cómo se usa en el análisis financiero?


Este código crea un modelo de regresión lineal usando datos de ejemplo (años e ingresos). Luego, predice los ingresos para el año 2024 y visualiza la relación entre los años y los ingresos.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: Año vs Ingresos (en miles de dólares)
años = np.array([2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]).reshape(-1, 1)
ingresos = np.array([500, 600, 700, 800, 900,1000,1500,2000])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(años, ingresos)

# Predicción
año_nuevo = np.array([2024]).reshape(-1, 1)
ingreso_predicho = modelo.predict(año_nuevo)

print(f"Para el año 2024, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil dólares")

# Visualización
plt.scatter(años, ingresos, color='blue')
plt.plot(años, modelo.predict(años), color='red')
plt.xlabel('Año')
plt.ylabel('Ingresos (mil dólares)')
plt.title('Regresión Lineal: Año vs Ingresos')
plt.show()
```
