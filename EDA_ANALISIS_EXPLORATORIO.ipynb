```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis'],
    'Edad': [23, 25, 23, 30, 25],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Sevilla', 'Barcelona'],
    'Salario': ['$3000', '$3500', None, '$4000', '$3500'],
    'Fecha_Contratacion': ['2020-01-15', '2019-11-23', '2020-01-15', '2018-05-30', '2019-11-23']
}

df = pd.DataFrame(data)

# Mostrar el DataFrame original
print("DataFrame Original:")
print(df)

# 1. Eliminación de duplicados
df = df.drop_duplicates()

# 2. Manejo de valores faltantes
df['Salario'] = df['Salario'].fillna('$0')

# 3. Corrección de errores
df['Ciudad'] = df['Ciudad'].str.lower()

# 4. Conversión de tipos de datos
df['Fecha_Contratacion'] = pd.to_datetime(df['Fecha_Contratacion'])
df['Salario'] = df['Salario'].str.replace('$', '').astype(float)

# Mostrar el DataFrame limpio
print("\nDataFrame Limpio:")
print(df)

# Análisis Exploratorio de Datos (EDA)
# Descriptiva de datos
print("\nDescriptiva de Datos:")
print(df.describe())

# Visualización de datos
plt.figure(figsize=(10, 6))
sns.histplot(df['Salario'], kde=True)
plt.title('Distribución de Salarios')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.show()

# Detección de outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Salario'])
plt.title('Boxplot de Salarios')
plt.xlabel('Salario')
plt.show()

```
