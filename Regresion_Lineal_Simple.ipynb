```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: horas estudiadas vs calificación
horas_estudio = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
calificaciones = np.array([2, 4, 5, 4, 5])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predicción
horas_nuevas = np.array([6]).reshape(-1, 1)
calificacion_predicha = modelo.predict(horas_nuevas)

print(f"Si estudias 6 horas, tu calificación predicha es: {calificacion_predicha[0]:.2f}")

# Visualización
plt.scatter(horas_estudio, calificaciones, color='blue')
plt.plot(horas_estudio, modelo.predict(horas_estudio), color='red')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: tamaño de la vivienda (m2) vs precio (en miles de dólares)
tamaño_vivienda = np.array([50, 60, 80, 100, 120]).reshape(-1, 1)
precio_vivienda = np.array([150, 180, 220, 260, 300])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(tamaño_vivienda, precio_vivienda)

# Predicción
tamaño_nuevo = np.array([90]).reshape(-1, 1)
precio_predicho = modelo.predict(tamaño_nuevo)

print(f"Para una vivienda de 90 m2, el precio predicho es: ${precio_predicho[0]:.2f} mil dólares")

# Visualización
plt.scatter(tamaño_vivienda, precio_vivienda, color='blue')
plt.plot(tamaño_vivienda, modelo.predict(tamaño_vivienda), color='red')
plt.xlabel('Tamaño de Vivienda (m2)')
plt.ylabel('Precio de Vivienda (mil dólares)')
plt.title('Regresión Lineal: Tamaño de Vivienda vs Precio')
plt.show()
```
