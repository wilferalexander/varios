**🔍📈 ¿Qué es la regresión lineal y cómo se usa en el área de la salud?**

La regresión lineal es una técnica de machine learning que nos ayuda a entender y predecir la relación entre dos variables. En el área de la salud, puede ser una herramienta poderosa para predecir y analizar diversos aspectos médicos.

¿Cómo funciona?

**Datos de entrada:**
 Tienes datos históricos, como la relación entre el índice de masa corporal (IMC) y la presión arterial.

**Dibuja una línea:**
La regresión lineal encuentra la mejor línea recta que atraviesa esos puntos. Esta línea se llama "línea de mejor ajuste".

**Predicción:**
Puedes usar esta línea para predecir resultados. Por ejemplo, si sabes el IMC de una persona, puedes predecir su presión arterial.
Fórmula mágica: La fórmula de la línea de regresión es: Y = mX + b

Y es el valor que queremos predecir (presión arterial).
X es el valor de entrada (IMC).
m es la pendiente de la línea (indica cuánto cambia Y por cada cambio en X).
b es la intersección con el eje Y (el valor de Y cuando X es 0).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: IMC vs Presión Arterial (en mm Hg)
imc = np.array([18.5, 24.0, 27.5, 30.0, 35.0]).reshape(-1, 1)
presion_arterial = np.array([120, 130, 140, 150, 160])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(imc, presion_arterial)

# Predicción
imc_nuevo = np.array([28.0]).reshape(-1, 1)
presion_predicha = modelo.predict(imc_nuevo)

print(f"Para un IMC de 28.0, la presión arterial predicha es: {presion_predicha[0]:.2f} mm Hg")

# Visualización
plt.scatter(imc, presion_arterial, color='blue')
plt.plot(imc, modelo.predict(imc), color='red')
plt.xlabel('IMC')
plt.ylabel('Presión Arterial (mm Hg)')
plt.title('Regresión Lineal: IMC vs Presión Arterial')
plt.show()
```

Este código crea un modelo de regresión lineal usando datos de ejemplo (IMC y presión arterial).
 Luego, predice la presión arterial para una persona con un IMC de 28.0 y visualiza la relación entre el IMC y la presión arterial.

¡Y eso es todo! Con la regresión lineal, puedes hacer predicciones simples pero poderosas en el área de la salud. 🚀
