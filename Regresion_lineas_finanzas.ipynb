Este c칩digo crea un modelo de regresi칩n lineal usando datos de ejemplo (a침os e ingresos). Luego, predice los ingresos para el a침o 2024 y visualiza la relaci칩n entre los a침os y los ingresos.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: A침o vs Ingresos (en miles de d칩lares)
a침os = np.array([2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]).reshape(-1, 1)
ingresos = np.array([500, 600, 700, 800, 900,1000,1500,2000])

# Crear el modelo de regresi칩n lineal
modelo = LinearRegression()
modelo.fit(a침os, ingresos)

# Predicci칩n
a침o_nuevo = np.array([2024]).reshape(-1, 1)
ingreso_predicho = modelo.predict(a침o_nuevo)

print(f"Para el a침o 2024, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil d칩lares")

# Visualizaci칩n
plt.scatter(a침os, ingresos, color='blue')
plt.plot(a침os, modelo.predict(a침os), color='red')
plt.xlabel('A침o')
plt.ylabel('Ingresos (mil d칩lares)')
plt.title('Regresi칩n Lineal: A침o vs Ingresos')
plt.show()
```
