## 🔍📈 ¿Qué es la regresión lineal y cómo se usa en ciencias sociales?

La regresión lineal es una técnica de machine learning que nos ayuda a entender y predecir la relación entre dos variables. En las ciencias sociales, puede ser una herramienta poderosa para analizar la relación entre variables como el ingreso y el nivel educativo en diferentes poblaciones.

¿Cómo funciona?

**Datos de entrada:**
Tienes datos históricos, como el nivel educativo y el ingreso de diferentes personas.

**Dibuja una línea:**
 La regresión lineal encuentra la mejor línea recta que atraviesa esos puntos. Esta línea se llama "línea de mejor ajuste".

**Predicción:**
Puedes usar esta línea para predecir resultados futuros. Por ejemplo, si sabes el nivel educativo de una persona, puedes predecir su ingreso.
Fórmula mágica: La fórmula de la línea de regresión es: Y = mX + b

Y es el valor que queremos predecir (ingreso).
X es el valor de entrada (nivel educativo).
m es la pendiente de la línea (indica cuánto cambia Y por cada cambio en X).
b es la intersección con el eje Y (el valor de Y cuando X es 0).


Este código crea un modelo de regresión lineal usando datos de ejemplo (nivel educativo e ingreso). Luego, predice el ingreso para una persona con 22 años de educación y visualiza la relación entre el nivel educativo y el ingreso.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo: Nivel Educativo (años de educación) vs Ingreso (en miles de dólares)
nivel_educativo = np.array([5, 8, 10, 12, 14, 16, 18, 20]).reshape(-1, 1)
ingreso = np.array([20, 30, 35, 40, 50, 60, 70, 80])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(nivel_educativo, ingreso)

# Predicción
nivel_educativo_nuevo = np.array([22]).reshape(-1, 1)
ingreso_predicho = modelo.predict(nivel_educativo_nuevo)

print(f"Para un nivel educativo de 22 años, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil dólares")

# Visualización
plt.scatter(nivel_educativo, ingreso, color='blue')
plt.plot(nivel_educativo, modelo.predict(nivel_educativo), color='red')
plt.xlabel('Nivel Educativo (años)')
plt.ylabel('Ingreso (mil dólares)')
plt.title('Regresión Lineal: Nivel Educativo vs Ingreso')
plt.show()
```

**Ejemplo considerando también la experiencia laboral:**

Este código amplía el modelo de regresión lineal para incluir tanto el nivel educativo como la experiencia laboral como factores predictivos del ingreso.

```python
# Datos de ejemplo: Nivel Educativo (años de educación), Experiencia Laboral (años) vs Ingreso (en miles de dólares)
nivel_educativo = np.array([5, 8, 10, 12, 14, 16, 18, 20])
experiencia_laboral = np.array([1, 3, 5, 7, 10, 12, 15, 20])
ingreso = np.array([20, 30, 35, 40, 50, 60, 70, 80])

# Crear una matriz de características
X = np.column_stack((nivel_educativo, experiencia_laboral))

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, ingreso)

# Predicción
nuevo_educacion_experiencia = np.array([[22, 25]])
ingreso_predicho = modelo.predict(nuevo_educacion_experiencia)

print(f"Para un nivel educativo de 22 años y 25 años de experiencia laboral, el ingreso predicho es: ${ingreso_predicho[0]:.2f} mil dólares")

# Visualización
plt.scatter(nivel_educativo, ingreso, color='blue')
plt.plot(nivel_educativo, modelo.predict(X), color='red')
plt.xlabel('Nivel Educativo (años)')
plt.ylabel('Ingreso (mil dólares)')
plt.title('Regresión Lineal: Nivel Educativo y Experiencia vs Ingreso')
plt.show()
```
