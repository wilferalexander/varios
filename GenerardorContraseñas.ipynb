```python
import secrets
import string
from IPython.display import display, HTML
from ipywidgets import widgets
from IPython.display import Javascript

def generate_password(length=16):
    """
    Genera una contraseña aleatoria.

    Args:
        length (int): Longitud de la contraseña. Por defecto es 16.

    Returns:
        str: Contraseña generada.
    """
    if length <= 0:
        raise ValueError("La longitud de la contraseña debe ser mayor que cero.")

    alphabet = string.ascii_letters + string.digits + string.punctuation
    return ''.join(secrets.choice(alphabet) for _ in range(length))

# Crear widgets
length_input = widgets.IntText(
    value=16,
    description='Longitud:',
    min=1,
    max=100
)

password_output = widgets.Text(
    value='',
    description='Contraseña:',
    disabled=True,
    style={'description_width': 'initial'}
)

generate_button = widgets.Button(
    description='Generar Contraseña',
    button_style='primary'
)

copy_button = widgets.Button(
    description='Copiar Contraseña',
    button_style='info',
    disabled=True  # Inicialmente deshabilitado hasta que se genere una contraseña
)

status_label = widgets.HTML(
    value="<p style='color: gray;'>Esperando generar contraseña...</p>"
)

def copy_to_clipboard(text):
    # Crear el JavaScript para copiar al portapapeles
    js_code = f"""
    async function copyToClipboard() {{
        try {{
            await navigator.clipboard.writeText("{text}");
            document.querySelector("#status_message").innerHTML =
                "<p style='color: green;'>¡Contraseña copiada al portapapeles!</p>";
        }} catch (err) {{
            document.querySelector("#status_message").innerHTML =
                "<p style='color: red;'>Error al copiar al portapapeles</p>";
            console.error('Error al copiar:', err);
        }}
    }}
    copyToClipboard();
    """
    return Javascript(js_code)

def on_generate_button_click(b):
    try:
        password = generate_password(length_input.value)
        password_output.value = password
        status_label.value = "<p id='status_message' style='color: green;'>¡Contraseña generada exitosamente!</p>"
        copy_button.disabled = False  # Habilitar el botón de copiar
    except ValueError as e:
        status_label.value = f"<p id='status_message' style='color: red;'>Error: {str(e)}</p>"
        copy_button.disabled = True  # Deshabilitar el botón de copiar

def on_copy_button_click(b):
    if password_output.value:
        display(copy_to_clipboard(password_output.value))

# Conectar los botones con sus funciones
generate_button.on_click(on_generate_button_click)
copy_button.on_click(on_copy_button_click)

# Mostrar la interfaz
display(widgets.VBox([
    widgets.HBox([length_input]),
    widgets.HBox([password_output]),
    widgets.HBox([generate_button, copy_button]),
    status_label
]))
```
